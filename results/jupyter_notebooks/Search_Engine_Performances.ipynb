```python
wolfram = [92.8,46.3,0,0,0,0]
google = [94.3,53.4,9.2,0,0,0]
ptgq = [92.8,81.8,76.3,86.3,90.7,85.7]
ptgq_tree = [92.8,90.7,85,90.2,90.7,100]
complexity = [1,2,3,4,5,6]
```

```python
fig=plt.figure(figsize=(14,7), facecolor='w', edgecolor='k')
plt.plot(complexity, wolfram ,color="brown", label='Wolfram', linewidth=2, linestyle='dotted')
plt.plot(complexity, google ,color="green", label='Google', linewidth=2, linestyle='dashed')
plt.plot(complexity, ptgq ,color="blue", label='PTGQ', linewidth=2, )
plt.plot(complexity, ptgq_tree ,color="red", label='PTGQ(only query tree)', linewidth=2, linestyle='dashdot')
# 
plt.xlabel('Query Complexity',fontsize=20)
plt.ylabel('Percentage of answered queries',fontsize=20)
# plt.title('DP + QT Gen Time v/s Complexity',fontsize=13)
plt.ylim(-1, 100.0)
# plt.xticks(np.arange(0,9,1), np.arange(0,9,1))
plt.rc('ytick', labelsize=16)
plt.rc('xtick', labelsize=16)
plt.legend(prop={'size':15})
plt.show() 
```
